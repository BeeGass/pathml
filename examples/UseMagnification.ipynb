```python
wsi = HESlide("/Volumes/sh/Glioma_IDH/svs/189674.svs", name = "example")
```

```python
# How big is the slide at different levels (resolutions)?
for i in range(3):
    print(f"level {i}:\t{wsi.slide.get_image_shape(level = i)}")
```

```python
# Look at some specific regions at different magnifications
region0 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),magnification='2.5x')
region1 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),magnification='5x')
region2 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),magnification='10x')
region3 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),magnification='20x')
region4 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),magnification='40x')
region = [region0,region1,region2,region3,region4]                               
title = ['2.5x','5x','10x','20x','40x']

fig = plt.figure(figsize=(8,8))
a = 1
for r in range(3):
    for c in range(2):
        fig.add_subplot(3,2,a)
        plt.axis('off')
        if a<=5:
            plt.title(title[a-1],fontsize=18)
            plt.imshow(region[a-1])
        a = a+1
        
fig.tight_layout()
```

```python
# Look use level to tile
region0 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),level=0)
region1 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),level=1)
region2 = wsi.slide.extract_region(location = (20000, 20000), size = (256, 256),level=2)
region = [region0,region1,region2]                               
title = ['level0','level1','level2']

fig = plt.figure(figsize=(8,8))
a = 1
for c in range(3):
    fig.add_subplot(1,3,a)
    plt.axis('off')
    plt.title(title[a-1],fontsize=18)
    plt.imshow(region[a-1])
    a = a+1
        
fig.tight_layout()
```
